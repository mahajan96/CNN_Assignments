image function does the segregation of arrays for images of different size and storing pose , expression and accessory values accordingly

```python
def image(root):
    y1_pose_label = []
    y2_pose_label = []
    y4_pose_label = []

    y1_expression_label = []
    y2_expression_label = []
    y4_expression_label = []

    y1_accessory_label = []
    y2_accessory_label = []
    y4_accessory_label = []

    a1 = []
    a2 = []
    a4 = []

    npath = None
    for path, directories, files in os.walk(root):
        for name in files:
            npath = (os.path.join(path, name)).replace(os.sep, '/')
            if npath[-5].isdigit() and npath[-5] == '2':
                imga2 = pgm2pil.imageOpenWrapper(npath)
                imga2 = array(imga2)
                words = npath.split("_")

                a2.append(imga2)
                pose_label_index = Pose.index(words[1])
                y2_pose_label.append(pose_label_index)
                expression_label_index = Expression.index(words[2])
                y2_expression_label.append(expression_label_index)
                accessory_label_index = Accessory.index(words[3])
                y2_accessory_label.append(accessory_label_index)

            elif npath[-5].isdigit() and npath[-5] == '4':
                imga4 = pgm2pil.imageOpenWrapper(npath)
                imga4 = array(imga4)
                words = npath.split("_")

                a4.append(imga4)
                pose_label_index = Pose.index(words[1])
                y4_pose_label.append(pose_label_index)
                expression_label_index = Expression.index(words[2])
                y4_expression_label.append(expression_label_index)
                accessory_label_index = Accessory.index(words[3])
                y4_accessory_label.append(accessory_label_index)


            else:

                if ".anonr" in npath:                                  
                    continue
                else:
                    imga1 = pgm2pil.imageOpenWrapper(npath)
                    imga1 = array(imga1)
                    words = npath.split("_")

                    a1.append(imga1)
                    pose_label_index = Pose.index(words[1])
                    y1_pose_label.append(Pose.index(words[1]))
                    expression_label_index = Expression.index(words[2])
                    y1_expression_label.append(expression_label_index)
                    accessory_label_index = Accessory.index(words[3].split('.')[0])
                    y1_accessory_label.append(accessory_label_index)
                    a1.append(imga1)

    a1 = np.array(a1)
    a2 = np.array(a2)
    a4 = np.array(a4)

    y1_pose_label = np.array(y1_pose_label)
    y2_pose_label = np.array(y2_pose_label)
    y4_pose_label = np.array(y4_pose_label)

    y1_expression_label = np.array(y1_expression_label)
    y2_expression_label = np.array(y2_expression_label)
    y4_expression_label = np.array(y4_expression_label)

    y1_accessory_label = np.array(y1_accessory_label)
    y2_accessory_label = np.array(y2_accessory_label)
    y4_accessory_label = np.array(y4_accessory_label)


    return a1, a2, a4, y1_pose_label, y1_expression_label, y1_accessory_label, y2_pose_label, y2_expression_label, y2_accessory_label, y4_pose_label, y4_expression_label, y4_accessory_label
```

```python
def train():
    a1, a2, a4, y1_pose_label, y1_expression_label, y1_accessory_label, y2_pose_label, y2_expression_label, y2_accessory_label, y4_pose_label, y4_expression_label, y4_accessory_label = image(
        root)

    shuffle_index = np.random.permutation(a4.shape[0])
    trainsize = int(a4.shape[0] * 0.8)


    a4_train, a4_test, acc_train, acc_test = a4[shuffle_index[:trainsize], :], a4[shuffle_index[trainsize:], :], \
                                         y4_accessory_label[shuffle_index[:trainsize]], y4_accessory_label[
                                             shuffle_index[trainsize:]]
    return a4_train, a4_test,acc_train ,acc_test
```

This class trains the data that we obtain from train function to classify between open and sunglasses to predict the testing values. Here we also find confusion matrix, the accuracy, the recall and the precision of our classiÔ¨Åer and then plot a ROC curve for both the sets Training and Testing.

```python
class open_vs_sunglasses:
#Calling train

    x_train ,x_test,y_train, y_test = train()

    nsamples, nx, ny = x_train.shape
    x_train = x_train.reshape((nsamples,nx*ny))

    nsamples, nx, ny = x_test.shape
    x_test= x_test.reshape((nsamples,nx*ny))
    
# Now checking if open or sunglasses
    
    
    y_training = (y_train == 0)
    y_testing = (y_test == 0)

    lg = LogisticRegression(max_iter = 10)

    lg.fit(x_train,y_training)

    y_pred = lg.predict(x_train)
    print('Classifier accuracy on the training set is {0} '.format(accuracy_score(y_training, y_pred)))


    print('Precision is {0} '.format(precision_score(y_training, y_pred)))

    print( 'And the confusion matrix: \n {0}'.format(confusion_matrix(y_training, y_pred) ))
    print('Recall is {0} '.format(recall_score(y_training, y_pred)))

    # Testing data
    y_pred_test = lg.predict(x_test)
    print('Classifier accuracy on the testing set is {0} '.format(accuracy_score(y_testing, y_pred_test)))


    print('Precision is {0} '.format(precision_score(y_testing, y_pred_test)))

    print( 'And the testing confusion matrix: \n {0}'.format(confusion_matrix(y_testing, y_pred_test) ))
    print('Recall is {0} '.format(recall_score(y_testing, y_pred_test)))

    #Roc curve

    from sklearn.model_selection import cross_val_predict

    y_scores = cross_val_predict(lg, x_train, y_training, cv=3, method="decision_function")
    fpr, tpr, thresholds = roc_curve(y_training,y_scores)

    def train_roc_curve(fpr, tpr, label=None):
        plt.plot(fpr, tpr, linewidth=2, label=label)
        plt.axis([0, 1, 0, 1])
        plt.xlabel('False Positive Rate', fontsize=16)
        plt.ylabel('True Positive Rate', fontsize=16)
    plt.figure(figsize=(8, 6))
    train_roc_curve(fpr, tpr)
    plt.show()

    print("testing roc ")
    y_scores1 = cross_val_predict(lg, x_test, y_testing, cv=3, method="decision_function")
    fpr, tpr, thresholds = roc_curve(y_testing,y_scores1)

    def train_roc_curve(fpr, tpr, label=None):
        plt.plot(fpr, tpr, linewidth=2, label="testing")
        plt.axis([0, 1, 0, 1])
        plt.xlabel('False Positive Rate', fontsize=16)
        plt.ylabel('True Positive Rate', fontsize=16)
    plt.figure(figsize=(8, 6))
    train_roc_curve(fpr, tpr)
    plt.show()
```
